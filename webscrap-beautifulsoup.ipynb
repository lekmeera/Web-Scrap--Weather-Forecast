```python
from bs4 import BeautifulSoup
import urllib.request
import bs4 as bs
import pandas as pd
```

```python
#opening url

sause = urllib.request.urlopen('https://en.tutiempo.net/climate/01-2013/ws-433710.html').read() 
soup = bs.BeautifulSoup(sause,'lxml')
```

```python
print(soup.prettify())
```

```python
print(soup.title.text)
```

```python
table = soup.find_all('table')
```

```python
print(table)
```

```python
My_table = soup.find('table',{'class':'medias mensuales numspan'})
```

```python
print(My_table)
```

```python
table_body=soup.find('tbody')
rows = My_table.find_all('tr')
for row in rows:
    cols=row.find_all('td')
    cols=[x.text.strip() for x in cols]
    print (cols)
```
